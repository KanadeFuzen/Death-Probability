```python
df = df[~df['Year'].isin(['Footnote','The code is C.'])]
df = df.set_index('Year')
df = df.drop(columns=['Source Year', 'Value Footnotes'])
df = df.rename(columns={'Country or Area': 'Country', 'Cause of death (WHO)': 'Death_Cause', 'Record Type': 'Record'})
df.head()
```

```python
target = 'Death_Cause'
y = df[target]
X = df.drop(columns=target)

mask1 = X.index <= '2016'
mask2 = X.index == '2017'
mask3 = X.index >= '2018'
X_train, y_train = X.loc[mask1], y.loc[mask1]
X_val, y_val = X.loc[mask2], y.loc[mask2]
X_test, y_test = X.loc[mask3], y.loc[mask3]

print("Baseline Accuracy = ", y.value_counts(normalize=True).max())
print("Baseline Accuracy From Train = ", y.train_counts(normalize=True).max())
print("Baseline Accuracy From ROC = ", roc_auc_score(y_train, [0.2]*len(y_train)))
```
